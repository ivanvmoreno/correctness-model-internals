```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('/home/arnau/Desktop/Cambridge/MARS/correctness-model-internals/evaluations/mistral_7b_instruct/math_operations_6k/base/main_generations_evaluated.csv')
df.head()
```

```python
size = 200
operations = ['sum', 'subtraction', 'multiplication', 'division']
for i, operation in enumerate(operations):
    print(f'Operation: {operation}')
    print("overall accuracy:", df.iloc[size*(i*7):size*(i*7 + 7)]['correct'].mean()*100)
    for difficulty in range(7):
        print(f'Difficulty: {difficulty}')
        print("Accuracy:", df.iloc[size*(i*7 + difficulty):size*(i*7 + (difficulty+1))]['correct'].mean()*100)
    print()
```
